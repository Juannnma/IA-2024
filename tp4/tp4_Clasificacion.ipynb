# ----------------------------------------------------
# 1. Entendimiento del problema y descripción del dataset
# ----------------------------------------------------

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
data = load_breast_cancer()

# Convertir a DataFrame
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

# Mostrar las primeras filas del dataset
df.head()
```

# ----------------------------------------------------
# 2. Exploración de los datos
# ----------------------------------------------------

```python

# Información del dataset
df.info()

# Estadísticas descriptivas
df.describe()
```

```python
# GRAFICO DE DISTRIBUCION DE LA VARIABLE OBJETIVO
plt.figure(figsize=(8, 5))
ax = sns.countplot(data=df, x='target')
plt.title('Distribución de tumores malignos y benignos')
plt.xlabel('Clase (0 = Maligno, 1 = Benigno)')
plt.ylabel('Cantidad')

# NUMEROS DE CONTEO ENCIMA DE LAS BARRAS
for p in ax.patches:
    ax.annotate(f'{p.get_height()}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom')

plt.show()
```

```python
# Correlación de variables
plt.figure(figsize=(15, 15))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Mapa de calor de las correlaciones entre características')
plt.show()
```


# ----------------------------------------------------
# 3. Preprocesamiento de los datos
# ----------------------------------------------------

```python


# Separar las características (X) de la variable objetivo (y)
X = df.drop('target', axis=1)
y = df['target']


```

```python
# Dividir el dataset en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Escalado de características
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# ----------------------------------------------------
# 4.  Entrenamiento del modelo
# ----------------------------------------------------

```python
# Crear y entrenar el modelo Random Forest
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```


# ----------------------------------------------------
# 5. Evaluación del modelo
# ----------------------------------------------------

```python
# Matriz de confusión
print("Matriz de confusión:")
print(confusion_matrix(y_test, y_pred))

# Reporte de clasificación
print("Reporte de clasificación:")
print(classification_report(y_test, y_pred))
```

# ----------------------------------------------------
# 6. Ajuste de hiperparámetros (Grid Search)
# ----------------------------------------------------

```python
# Definir parámetros para GridSearchCV
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}
# Búsqueda de hiperparámetros con validación cruzada
grid_search = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)
```

```python
# Resultados de la búsqueda
print("Mejores hiperparámetros encontrados:")
print(grid_search.best_params_)
```

```python
# Reentrenar el modelo con los mejores parámetros
best_model = grid_search.best_estimator_
best_model.fit(X_train, y_train)

# Evaluar el mejor modelo
y_pred_best = best_model.predict(X_test)
print("Reporte de clasificación del mejor modelo:")
print(classification_report(y_test, y_pred_best))
```

```python
# Curva ROC para el mejor modelo
y_pred_proba_best = best_model.predict_proba(X_test)[:, 1]
fpr_best, tpr_best, _ = roc_curve(y_test, y_pred_proba_best)
roc_auc_best = roc_auc_score(y_test, y_pred_proba_best)

plt.figure()
plt.plot(fpr_best, tpr_best, color='orange', lw=2, label=f'Curva ROC (AUC = {roc_auc_best:.2f})')
plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curva ROC (Mejor Modelo)')
plt.legend(loc="lower right")
plt.show()
```
